```python
from random import sample

cifras = 4
jugar = True
ganar = False
picas = 0
fijas = 0
intentos = 0

def generar_numero_random(cantidad_cifras):
    digitos = list(range(10))  # 0-9
    numero_random = sample(digitos, cantidad_cifras)
    return numero_random

def retornar_picas_fijas(numero_usuario, numero_secreto):
    fijas = 0
    picas = 0
    for digito in range(cifras):
        if numero_usuario[digito] in numero_secreto:
            if numero_usuario[digito] == numero_secreto[digito]:
                fijas = fijas + 1
            else:
                picas = picas + 1
    return picas, fijas

def dar_mensaje(intentos):
    if intentos <= 3:
        return "Excelente! Eres muy bueno en este juego."
    if intentos <= 6:
        return "Has adivinado el nmero."
    else:
        return "Por fin lo lograste"

# Generar nmero secreto al inicio
numero_secreto = generar_numero_random(cifras)
print("Picas y Fijas")
print("He generado un nmero de " + str(cifras) + " cifras")

while jugar:
    numero_jugador = input("Ingresa un nmero de " + str(cifras) + " cifras: ")
    cantidad_cifras = len(numero_jugador)

    if cantidad_cifras == cifras and numero_jugador.isdigit():
        # Convertir a lista de nmeros
        numero = list(numero_jugador)
        numero = [int(i) for i in numero]

        # Verificar si gana
        if numero == numero_secreto:
            ganar = True
            jugar = False
            print(" Adivinaste el nmero.")
            print(dar_mensaje(intentos))
        else:
            # Calcular picas y fijas
            picas, fijas = retornar_picas_fijas(numero, numero_secreto)
            intentos = intentos + 1

            print("Picas: " + str(picas) + ", Fijas: " + str(fijas))
            print("Intento nmero: " + str(intentos))

            # Lmite de intentos
            if intentos >= 10:
                print("PERDISTE! Se te acabaron los intentos.")
                print("El nmero era: " + "".join(str(x) for x in numero_secreto))
                jugar = False
    else:
        print("Debe ser un nmero de " + str(cifras) + " cifras")
        intentos = intentos + 1

print("juego terminado")
```
